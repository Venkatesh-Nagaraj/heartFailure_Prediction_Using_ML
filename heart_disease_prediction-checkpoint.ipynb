```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
cat_dict={'Yes':1,'No':0}    
```

```python
ran=[0,2,3,4,7,12,15,16,17]
```

```python
for i in ran:
    dataset.iloc[:,i]=dataset.iloc[:,i].apply(lambda x:cat_dict[x])
```

```python
dataset['Diabetic'].value_counts()
```

```python
plt.figure(figsize=(13,10))
sb.countplot(x=dataset['PhysicalHealth'],data=dataset)
plt.show()
```

```python
data_diabet=LabelEncoder()
data_sex=LabelEncoder()
data_age=LabelEncoder()
data_race=LabelEncoder()
data_gen=LabelEncoder()
data_phyh=LabelEncoder()
```

```python
dataset['Diabetic']=data_diabet.fit_transform(dataset['Diabetic'])
dataset['Sex']=data_sex.fit_transform(dataset['Sex'])
dataset['AgeCategory']=data_age.fit_transform(dataset['AgeCategory'])
dataset['Race']=data_race.fit_transform(dataset['Race'])
dataset['GenHealth']=data_gen.fit_transform(dataset['GenHealth'])
dataset['PhysicalHealth']=data_phyh.fit_transform(dataset['PhysicalHealth'])
```

```python
X_lab=dataset.loc[:,['Smoking','Stroke','DiffWalking','AgeCategory','Diabetic','PhysicalHealth','KidneyDisease']]
```

```python
target=dataset['HeartDisease']
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_lab, target ,random_state=25,train_size = .80)
```

```python
X_train.shape
```

## Train the Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=25)
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
from sklearn import tree
dt_model = tree.DecisionTreeClassifier(criterion="entropy")
```

```python
dt_model.fit(X_train,y_train)
```

```python
dt_model.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
```

```python
log_model.fit(X_test,y_test)
```

```python
log_model.score(X_test,y_test)
```

### Predict the model

```python
y_predicted = model.predict(X_test)
```
